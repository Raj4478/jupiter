## MATPLOTLIB

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
xpts = np.array([0, 10])
ypts = np.array([0, 250])

plt.plot(xpts, ypts)
plt.show()
```

```python
dataframe = pd.DataFrame({
    "Cricket_Bat": ["SG", "Kookaburra", "Gray Nicolls", "MRF", "Adidas"],
    "MRP": [5000, 6000, 7000, 8000, 9000],
    "WEIGHT_GRAM": [1000, 1200, 1100, 1300, 1400]
})

plt.plot(dataframe["Cricket_Bat"], dataframe["MRP"])
plt.grid()
```

```python
dataframe = pd.DataFrame({
    "Cricket_Bat": ["SG", "Kookaburra", "Gray Nicolls", "MRF", "Adidas"],
    "MRP": [5000, 6000, 7000, 8000, 9000],
    "WEIGHT_GRAM": [1000, 1200, 1100, 1300, 1400]
})

plt.plot(dataframe["WEIGHT_GRAM"], dataframe["MRP"], marker='o')

plt.title("Cricket Bat MRP vs Weight", loc='left', pad=20, fontsize=14, color='blue')
```

```python
a= np.arange(5)
b= [2, 4, 6, 8, 10]
c= [5, 6, 7, 8, 9]

fig = plt.figure()
ax = plt.subplot()

ax.plot(a, b, 'k--', label='Frequency')
ax.plot(a, c, 'k:', label='Periods')

ax.legend()

plt.title("Line styles", fontsize=14)
```
