```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
penguins.species.value_counts()
```

```python
sns.set_style("ticks")
sns.set_context("paper")
sns.scatterplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue = "island", palette="Set3")
```

```python
sns.scatterplot(data=penguins, x="species", y="body_mass_g", hue="island", style="sex", alpha=0.5)
```

```python
sns.stripplot(data=penguins, x="species", y="body_mass_g", hue="island", dodge=True, jitter=True)
```

```python
sns.swarmplot(data=penguins, x="species", y="body_mass_g", hue="island", dodge=True)
```

```python
sns.set_context("notebook")
sns.histplot(data=penguins, x="body_mass_g", hue="species", multiple="stack")
```

```python
sns.regplot(data=penguins, x="flipper_length_mm", y="body_mass_g", marker="o")
```

```python
sns.jointplot(data=penguins, x="flipper_length_mm", y="body_mass_g", kind="reg")
```

```python
sns.barplot(data=penguins, x="species", y="body_mass_g", hue="island")
```

```python
sns.countplot(data=penguins, x="species", hue="island")
```
